```python
import numpy as np
import os
import pandas as pd
import cv2
import json
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
labels = os.listdir("pump/labels copy")
```

```python
for label in labels:
    with open("pump/labels copy/"+label) as f:
        lines = f.readlines()
        lines = [line.split() for line in lines]
        # replace 60 to 81 
        for line in lines:
            if line[0] == "81":
                line[0] = "80"

    with open("pump/labels copy/"+label, "w") as f:
        for line in lines:
            f.write(" ".join(line))
            f.write("\n")
        #pprint(lines)


    print(lines)        
    
```
